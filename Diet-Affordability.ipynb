```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Loading world map shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Merging data with the world map data using the 'Entity' column
merged_data = world.merge(df, left_on='name', right_on='Entity')

# Plotting
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
merged_data.plot(column='Share of the population who cannot afford a healthy diet', 
                 cmap='YlOrRd', 
                 linewidth=0.8, 
                 ax=ax, 
                 edgecolor='0.8', 
                 legend=True)

# Adding name as a text annotation
ax.text(x=0.00, y=-0.40, s='Representation of "Share of the population who cannot afford a healthy diet" using Python by @Muhammad_Danish', transform=ax.transAxes, fontsize=12, color='black')

ax.set_title('Share of the Population Unable to Afford a Healthy Diet (2017-2021)')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Show the plot
plt.show()
```
